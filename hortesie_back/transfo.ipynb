```python
def split2(string):
    if type(string)==str:
        return(string.split('-')[0])

def split3(string):
    if type(string)==str:
        if len(string.split('-'))>1:
            return(string.split('-')[1])

def split4(string):
    if type(string)==str:
        if len(string.split('-'))>2:
            return(string.split('-')[2])
```

```python
e = 'ezrrzr - rzaezer'
split2(e)
```

```python
surveys_df['pays']= surveys_df.apply(lambda row: split4(row['nom_fr']), axis=1)
```
